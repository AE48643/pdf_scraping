```python
#Importing necessary files
import openpyxl as op


# Load the workbook and the active sheet
excel_file_directory = "../pdf_file/DA 25.xlsx"
workbook = op.load_workbook(excel_file_directory)
ws = workbook.active

#Creating and calling new sheet to save all the new collected information
workbook.create_sheet("Summary")
summary_sheet = workbook['Summary']


def Credits_Summary (CreditCatagory) :
#Initial Variables
    creditType = CreditCatagory
    total_Credit = 0
    total_Credit_courses = 0

#Initial Variable for the Condition to Check if the Cell is available to print
    column_condition = False
    column_number = 1

    if creditType == "InP":
        for row in ws.iter_rows(min_row=1, max_row=ws.max_row):
            cell_n = row[13]  # Column N (index 13 because it's 0-based)
            cell_o = row[14]  # Column O (index 14)

            if cell_n.value in ['In P', 'In'] or cell_o.value in ['In P', 'In']:
                row_num = cell_n.row
                val_f = ws.cell(row=row_num, column=6).value  # Column F
                val_e = ws.cell(row=row_num, column=5).value  # Column E

# Add F value to total only if it's a number
                if isinstance(val_f, (int, float)): #and val_f < 10:
                    total_Credit += val_f
                    total_Credit_courses += 1

                
                print(f"Row {row_num}: Value in F = {val_f}, Value in E = {val_e}")

#Collecting info on Column N and O
    for row in ws.iter_rows(min_row=1, max_row=ws.max_row):
        cell_n = row[13]  # Column N (index 13 because it's 0-based)
        cell_o = row[14]  # Column O (index 14)

#Collecting the row number that have the credityType  
        if cell_n.value == creditType or cell_o.value == creditType:
            row_num = cell_n.row
#Using the row number outputing the value in f or e respective column            
            val_f = ws.cell(row=row_num, column=6).value  # Column F
            val_e = ws.cell(row=row_num, column=5).value  # Column E

# Add F value to total only if it's a number
            if isinstance(val_f, (int, float)): #and val_f < 10:
                total_Credit += val_f
                total_Credit_courses += 1

      
            print(f"Row {row_num}: Value in F = {val_f}, Value in E = {val_e}")

#Condition to Check if the Cell is available to print
    while column_condition is False:
        column_number += 1
        check_cell = summary_sheet.cell(row = 1, column = column_number)
        if check_cell.value is None:
            column_condition = True
        
#Double Checking if the condition is met then printing out the info
    if column_condition is True:

        c1 = summary_sheet.cell(row= 1, column=column_number)
        c1.value = f'{creditType}'

        c2 = summary_sheet.cell(row= 2, column=column_number)
        c2.value = total_Credit_courses

        c3 = summary_sheet.cell(row= 3, column=column_number)
        c3.value = total_Credit


        

#Seeing on the Console if it worked 
    print(f"Total {creditType} Credits: {total_Credit}")
    print(f"Total Completed Courses: {total_Credit_courses}")


#Checking for the type of course (wheter they are complete (C), InProgress (InP), Manual Map Completed (MC)...)
# Run summaries for each credit type
credit_types = ['C', 'MC', 'R', 'MR', 'T', 'TC', 'InP', 'E']
for ct in credit_types:
    Credits_Summary(ct)

#Defining the table titles
courses_title = summary_sheet.cell(row= 2, column=1)
courses_title.value = f'Total Couses'

credits_title = summary_sheet.cell(row= 3, column=1)
credits_title.value = f'Total Credits'

#Saving all the newly added info and sheet into a new excel file
workbook.save('../Final Product/Summary Audit.xlsx')
```
